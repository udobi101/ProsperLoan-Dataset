## Question 7

### what is the total number and percentage of missing data in 2014 to 3 decimal places?


```python
# missing values
missing_count = df['Y2014'].isnull().sum()
missing_count

# total points of the year 2014
total_no = df['Y2014'].shape[0]
total_no

# Calculating the percentage

missing_per = (missing_count/total_no)* 100

# print the results 

print("Total number of missing data: ", missing_count)
print("Percentage of missing data: ", round(missing_per, 3), "%")
```

## Question 12

### What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
mean_2017 = df['Y2017'].mean()
std_2017  = df['Y2017'].std()

# # Rounding the results to 2 decimal places

mean_2017 = round(mean_2017, 2)
std_2017 = round(std_2017, 2)

print("mean for the year 2017: ", mean_2017)
print("standard deviation of the year 2017: ", std_2017)
```
