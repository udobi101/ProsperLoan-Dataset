```python
#States with more Loans
base_color = sb.color_palette()[9]
ordered = df1['BorrowerState'].value_counts().index
plt.figure(figsize=[15, 10])
sb.countplot(data=df1,y='BorrowerState',color=base_color , order= ordered);
plt.title('Borrower State Distribution');
plt.xticks(rotation=90);
```

California, Florida and New York has the three biggest amount of loan.

```python
df1.groupby("EmploymentStatus")["TotalProsperLoans"].mean().sort_values(ascending = False)
```

Retired people tends to take more loans than other employment status, this could be that as they retire , thend to take more loans to offset their bills
